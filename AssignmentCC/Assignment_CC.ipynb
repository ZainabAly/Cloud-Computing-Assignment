```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/books.csv")
df
```

Data Preprocessing

```python
#identify missing values
print("\nMissing values:")
print(df.isnull().sum())
```

```python
#Fill missing values in this column with the most frequent value
df['isbn'].fillna(df['isbn'].mode()[0], inplace=True)
```

```python
#Fill missing values in this column with the most frequent value
df['isbn13'].fillna(df['isbn13'].mode()[0], inplace=True)
```

```python
#Fill missing values in this column with the most frequent value
df['original_publication_year'].fillna(df['original_publication_year'].mode()[0], inplace=True)
```

```python
#Fill missing values in this column with the most frequent value
df['original_title'].fillna(df['original_title'].mode()[0], inplace=True)
```

```python
#Fill missing values in this column with the most frequent value
df['language_code'].fillna(df['language_code'].mode()[0], inplace=True)
```

```python
#identify missing values after filling NA
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
df
```

```python
# Save the cleaned DataFrame to the original CSV file
df.to_csv('books.csv', index=False)
```

Harry Potter books analysis

```python
# Filter the dataset for titles containing "Harry Potter"
harry_potter_books = df[df['title'].str.contains('Harry Potter', case=False)]
```

```python
# Extract rating columns (assuming they are named rating1, rating2, rating3, rating4, rating5)
rating_columns = ['rating_1', 'rating_2', 'rating_3', 'rating_4', 'rating_5']
rating_columns
```

```python
# Calculate the average rating for each Harry Potter book
average_ratings = harry_potter_books.groupby('title')[['ratings_1', 'ratings_2', 'ratings_3', 'ratings_4', 'ratings_5']].mean()

# Display the average rating for each Harry Potter book
print("Average Ratings of Harry Potter Books:")
print(average_ratings)
```

```python
# Find the book with the highest sales (prints all info about the book)
top_selling_book = harry_potter_books.loc[harry_potter_books['average_rating'].idxmax()]

# Display the top-selling book
print("Top Selling Book within the Harry Potter Series:")
print(top_selling_book)
```
